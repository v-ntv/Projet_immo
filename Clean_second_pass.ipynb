```python
import pandas as pd
```

```python
df_error_ma_1 = pd.read_csv('error_ma_1.csv')
df_pdll_v2 = pd.read_csv('pays_de_la_loire_V2.csv')
```

```python
# 1. Obtenir les URLs uniques du premier DataFrame
urls_a_filtrer = df_error_ma_1['url'].unique()

# 2. Filtrer le second DataFrame en utilisant isin()
df_filtre = df_pdll_v2[df_pdll_v2['url'].isin(urls_a_filtrer)]

#Enlever les majuscules
df_filtre['url'] = df_filtre['url'].str.lower()

# Afficher le résultat
df_filtre
```

```python
# Vérifier si toutes les URLs sont en minuscules
est_tout_minuscule = df_filtre['url'].str.islower().all()

print(f"Toutes les URLs sont-elles en minuscules ? {est_tout_minuscule}")
```

```python
df_filtre.to_csv('villes_error_1.csv', index=False)
```
