```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.read_csv("DS_RP_POPULATION_COMP_2022_data.csv",sep=";")
df2 = pd.read_csv("DS_RP_POPULATION_COMP_2022_metadata.csv",sep=';')
```

```python
df_mapping_age = df2[df2['COD_VAR']=='AGE']
df_mapping_sex = df2[df2['COD_VAR']=='SEX']
df_mapping_pcs = df2[df2['COD_VAR']=='PCS']
df_mapping_time_period = df2[df2['COD_VAR']=='TIME_PERIOD']
df_mapping_rp_measure = df2[df2['COD_VAR']=='RP_MEASURE']
df_mapping_geo = df2[df2['COD_VAR']=='GEO']
```

```python
#Créer le dictionnaire pour le mapping
mapping = dict(zip(df2['COD_MOD'], df2['LIB_MOD']))

# Appliquer le mapping
df1['AGE'] = df1['AGE'].map(mapping)
df1['PCS'] = df1['PCS'].map(mapping)
df1['RP_MEASURE'] = df1['RP_MEASURE'].map(mapping)
df1['SEX'] = df1['SEX'].map(mapping)

df1
```
